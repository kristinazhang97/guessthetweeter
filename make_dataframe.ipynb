```python
import json
from cleantext import *
import pandas as pd
import numpy as np
from recent_search_milo import most_popular_tweeters
```

```python
# create a word set
ws = get_word_set()

print(f"length of word set: {len(ws)}")

# get the columns (add id to word set)
cols = ws
cols.append("tweeter_id")
```

```python
# make a dataframe to hold all the data
df = pd.DataFrame(columns=cols)
```

```python
# iterate through the tweeters' data
for tweeter in tqdm(most_popular_tweeters.keys()):
    # add the tweet data to all_bows
    temp_df = pd.DataFrame(data=get_tweeter_vectors(str(tweeter), ws), columns=cols)
    # append it to the dataframe
    df = pd.concat((df, temp_df))
```
